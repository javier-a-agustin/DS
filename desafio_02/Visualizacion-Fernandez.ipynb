```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

# Dataset import

```python
suicide_rate_by_country = pd.read_csv('suicide_rate_by_country.csv')
print(suicide_rate_by_country.shape)
suicide_rate_by_country.head()
```

```python
df = suicide_rate_by_country
```

```python
male_young_suicides = df[(df.country == "Argentina") & (df.sex == "male") & (df.age == '15-24 years')][['country', "suicides_no", 'year', 'sex', 'population', 'suicides/100k pop']]
female_young_suicides = df[(df.country == "Argentina") & (df.sex == "female") & (df.age == '15-24 years')][['country', "suicides_no", 'year', 'sex', 'population', 'suicides/100k pop']]
```

```python
fig, ax = plt.subplots()
rects_1 = ax.bar(male_young_suicides.year, male_young_suicides.suicides_no, label="male")
rects_2 = ax.bar(female_young_suicides.year, female_young_suicides.suicides_no, label="female")
ax.set_title('Suicide rates in Argentina over time')
ax.set_ylabel('Count')
ax.set_xlabel('Year')
ax.legend()
fig.set_size_inches(18.5, 10.5, forward=True)
ax.bar_label(rects_1, padding=2)
ax.bar_label(rects_2, padding=2)
plt.show()
```

En el grafico 1 se puede observar el ratio de suicidio sobre los anios.
Se puede observar que para la edad seleccionada (15-24) el ratio de suicidios en hombres a partir del anio 2000 duplica al ratio de suicidio de mujeres.
Mientras que el ratio de suicidios en mujeres, si bien aumenta, se mantiene mas lineal.

```python
fig, ax = plt.subplots()
ax.scatter(male_young_suicides.population, male_young_suicides.suicides_no, c='red')
ax.set_title("population vs. suicides_no on male population")
ax.set_xlabel("population")
ax.set_ylabel("suicides_no")
fig.set_size_inches(18.5, 10.5, forward=True)
```

Se puede observar que, si bien tiene sentido que aumente el ratio de suicidios cuando aumenta la poblacion, se nota un notorio aumento de suicidios en un, relativamente, corto aumento de la poblacion

```python
sns.histplot(data=male_young_suicides, x="year", hue="suicides_no")
```

Similar al primer grafico, se observa la progresion en el ratio de suicidios a medida del tiempo.

```python
sns.lineplot(data=female_young_suicides, x="year", y="suicides/100k pop")
```
